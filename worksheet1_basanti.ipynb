````python
# • Solve the following system of Linear equation using Inverse Methods.

# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

# {Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
# • Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
# of Numpy}
A = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B = np.array([-1, -3, 9])

x = np.linalg.inv(A).dot(B)
x
````

````python
#4.2 Experiment: How Fast is Numpy
# 1. Element-wise Addition (Python vs Numpy)
import numpy as np
import time
# Python
a = list(range(1000000))
b = list(range(1000000))

start = time.time()
c = [a[i] + b[i] for i in range(1000000)]
print("Python Time:", time.time() - start)

# Numpy
a_np = np.arange(1000000)
b_np = np.arange(1000000)

start = time.time()
c_np = a_np + b_np
print("Numpy Time:", time.time() - start)
````

````python
#2. Element-wise Multiplication
# Python
start = time.time()
c = [a[i] * b[i] for i in range(1000000)]
print("Python Time:", time.time() - start)

# Numpy
start = time.time()
c_np = a_np * b_np
print("Numpy Time:", time.time() - start)
````

````python
#3. Dot Product
# Python
start = time.time()
dot_py = sum(a[i] * b[i] for i in range(1000000))
print("Python Time:", time.time() - start)

# Numpy
start = time.time()
dot_np = np.dot(a_np, b_np)
print("Numpy Time:", time.time() - start)
````
